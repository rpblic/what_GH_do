```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

```python
df= pd.read_table('./Pre_processed_data/merged_data_exceptshopping.csv', delimiter= ',', encoding='euc-kr')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis= 1, inplace= True)
```

```python
agg_func= {'U_AM': 'sum', 'U_CT': 'sum'}
```

```python
df= df.groupby(['ID', 'BIZ_UNIT']).agg(agg_func)
```

```python
df.head(10)
```

```python
df.reset_index(drop= False, inplace= True)
```

```python
BIZ_list= sorted(df['BIZ_UNIT'].unique().tolist())
```

```python
df[df['BIZ_UNIT']== 'B01']
```

```python
sorted(BIZ_list)
```

```python
df_merge= pd.DataFrame(columns= ['ID'])
for i in BIZ_list:
    df_toadd= df[df['BIZ_UNIT']== i]
    df_toadd.drop('BIZ_UNIT', axis= 1, inplace= True)
    df_toadd.columns= df_toadd.columns.map(lambda x: x+'_'+str(i))
    idname= 'ID_'+str(i)
    df_toadd.rename(columns= {idname: 'ID'}, inplace= True)
    df_merge= pd.merge(df_merge, df_toadd, how= 'outer', on= 'ID')
df_merge.info()
```

```python
df_merge.sort_values('ID')
```

```python
df_merge.fillna(0, inplace=True)
```

```python
df_merge.to_csv('groupby_ID_nonshopping.csv')
```
