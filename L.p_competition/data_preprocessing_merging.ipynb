### 현재 Demographic 데이터에서 주소의 경우는 우편번호가 기록되어 구체적인 정보를 알 수 없는 문제가 있음.
### 이에 지명 데이터를 추가시키는 작업을 실시하였음.

지명에 대한 데이터는 다음에서 가져와 적용하였음: (http://www.epost.go.kr/search/zipcode/areacdAddressDown.jsp)

```python
df_adr= pd.read_table('.\\areacd_rangeaddr_DB\\20171206_도로명범위.txt', delimiter='|', engine='python')
```

```python
df_adr.tail(10)
```

```python
df_adr['우편번호']=df_adr['우편번호'].map(lambda x: str(x).zfill(5)[:3])
```

```python
df_adr['우편번호']
```

```python
df_adr[['우편번호', '시도', '시군구']].tail(20)
```

```python
df_adr.drop_duplicates(['우편번호', '시도', '시군구'], inplace= True)
```

```python
df_adr= df_adr[['우편번호', '시도', '시군구']]
```

```python
df_adr.columns= ['HOM_PST_NO', 'PRVIN', 'DESTRICT']
```

```python
df_adr.head()
```
