```python
import pandas as pd

import numpy as np

from datetime import datetime

```

```python
df_raw= pd.read_csv('C://Studying//GrowthHackers//L.p_competition//Pre_processed_data/merged_dataframe_shopping.csv', encoding='euc-kr', delimiter=',')
```

```python
df_2= pd.read_pickle('C://Studying//GrowthHackers//L.p_competition/find_unknown.pkl')
```

```python
df_raw.drop('Unnamed: 0', axis= 1, inplace= True)
```

```python
df_raw.head(10)
```

```python
df_2.head(20)
```

```python
df_raw['dtime']= (df_raw['YEAR']*10000 + df_raw['MONTH']*100 + df_raw['DAY']).apply(str).apply(lambda x: x[:4]+'-'+x[4:6]+'-'+x[6:])
```

```python
df_raw['dtime']= df_raw['dtime'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
df_raw.drop(['YEAR', 'MONTH', 'DAY', 'HOUR'], axis=1, inplace=True)
```

```python
agg_func= {'PD_S_C': lambda x: x.nunique(), 'BR_C': lambda x: x.nunique(), 'BUY_AM': 'sum', 'BUY_CT': 'sum'}
```

```python
df_rst= df_raw.groupby(['ID', 'RCT_NO', 'dtime']).agg(agg_func)
```

```python
df_rst.reset_index(drop= False, inplace= True)
```

```python
df_rst.info()
```

```python
df_rst[df_rst.duplicated(['ID', 'RCT_NO', 'dtime'])]
```

```python
df_rst.isnull().all()
```

```python
df_rst.sort_values(['ID', 'dtime'], ascending= True, inplace= True)
```

```python
df_rst.head(20)
```

```python
df_rstrst= pd.Series()
for i in range(20001):
    df_toadd= df_rst[df_rst['ID']== i]
    df_past= df_toadd['dtime']
    if df_past.size>1:
        df_past= pd.concat([pd.Series([df_past.iloc[0]]), df_past.iloc[:-1]])
    df_rstrst= pd.concat([df_rstrst, df_past])
df_rstrst.size
```

```python
df_rstrst.reset_index(drop= True, inplace= True)
```

```python
df_rst.reset_index(drop= True, inplace= True)
```

```python
pd.to_datetime(df_rst['dtime'], box= False, exact= True).iloc[0]
```

```python
(df_rst['dtime'] - df_rstrst).days
```

```python
df_rst['DIFF_DATE']= ((df_rst['dtime']- df_rstrst)/np.timedelta64(1, 'D')).astype(int)
```
