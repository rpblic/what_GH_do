```python
import numpy as np
import pandas as pd
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA, KernelPCA
from sklearn.preprocessing import RobustScaler
```

```python
data= pd.read_pickle('./item_selected_std0.1_corr0.2.pkl')
```

```python
data.reset_index(drop= True, inplace= True)
```

```python
data.shape[0]
```

```python
data_scaled= pd.DataFrame(RobustScaler().fit_transform(data), columns= data.columns)
```

```python
data.head(30)
```

```python
data_scaled.head(30)
```

```python
km= KModes(n_clusters= 3, verbose= 1)
clusters= km.fit_predict(data_scaled)
```

```python
print(km.cluster_centroids_)
```

```python
cluster_series= pd.Series(data= clusters)
```

```python
data.loc[cluster_series==0, :]
```

```python
pca_model_frame= {}
km_model_frame= {}
components_frame= {}
clusters_frame= {}
for i in range(5, 61, 5):
    pca= PCA(n_components= i).fit(data_scaled)
    pca_model_frame[i]= pca
    components_frame[i]= pd.DataFrame(pca.components_, columns= data.columns)
    data_pca= pca.transform(data_scaled)
    print('ratio: {}, score: {}\n'.format(pca.explained_variance_ratio_, pca.score(data_scaled)))
    for j in range(2, 11, 2):
        km= KModes(n_clusters= j, verbose= 1)
        clusters= km.fit_predict(data_pca)
        cluster_series= pd.Series(data= clusters)
        km_model_frame[j]= km
        clusters_frame[j]= cluster_series
        print(km.cluster_centroids_)
        print('\n# of pts: ')
        print([data.loc[cluster_series==k, :].shape[0] for k in range(j)])
        print('\n')
```
